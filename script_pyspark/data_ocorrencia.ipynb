```python
# verifica se os valores das colunas 'codigo_ocorrencia' e 'codigo_ocorrencia1' ,2,3 e 4 são todos iguais
compara_codigo = (df_oc_bruto.selectExpr('codigo_ocorrencia', 
                         '''case when 
                                cast(codigo_ocorrencia as INT) == cast(codigo_ocorrencia1 as INT) AND 
                                cast(codigo_ocorrencia1 as INT) == cast(codigo_ocorrencia2 as INT) AND
                                cast(codigo_ocorrencia2 as INT) == cast(codigo_ocorrencia3 as INT) AND 
                                cast(codigo_ocorrencia3 as INT) == cast(codigo_ocorrencia4 as INT)
                                then 1 else 0 end as validacao
                            '''
                         )     
        )
compara_codigo.select('validacao').distinct().limit(5).show()
```
